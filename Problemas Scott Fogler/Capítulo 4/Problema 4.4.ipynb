# Problema 4.4
###  Cálculo de la caída de presión en un lecho empacado
"Elementos de Ingeniería de las reacciones químicas" ,Scott Fogler capitulo 4, página 183

Primero importaremos la librerías que nos ayudaran a resolver el problema (No olvides activar las librerías con **Shift+enter**):

```python
import numpy
import matplotlib.pyplot as plt
```

```python
# datos
P0 = 10  # atm
v0 = 104.4 / 0.413  # lbm/(h*ft3)
betha0 = 164.1 / (144 * 14.7)  # atm/ft
z_tabla = numpy.arange(0.0, 70.0, 10)  # ft
P_tabla = P0 * (1 - 2*betha0*z_tabla/P0) ** 0.5  # atm
v_tabla = v0 * P0 / P_tabla
```

```python
# Tabla E4-4.1 perfiles de P y v
print("_" * 33)
print('Perfiles de P y v'.center(33))
print("_" * 33)
print(" {:10} {:10} {:10} ".format("z (ft)", "P (atm)", "v (ft³/h)"))
print("_" * 33)
for i in range(len(z_tabla)):
    print(" {:10} {:10} {:10} ".format(round(z_tabla[i], 2),
                                       round(P_tabla[i], 2),
                                       round(v_tabla[i], 2)))
print("¯" * 33)

z = numpy.arange(0.0, 60.0, 0.1)  # ft
P = P0 * (1 - 2*betha0*z/P0) ** 0.5  # atm
v = v0 * P0 / P
```

```python
# Graficando resultados
fig, ax = plt.subplots()
ax.minorticks_on()
ax.plot(z, P, 'k', linewidth=1.1)
ax.set_xlabel('$z(ft)$')
ax.set_ylabel('$P \\ \\left( atm \\right)$')
ax.grid(which='BOTH', ls=':')
ax.grid(True)

fig1, ax1 = plt.subplots()
ax1.minorticks_on()
ax1.plot(z, v, 'k', linewidth=1.1)
ax1.set_xlabel('$z(ft)$')
ax1.set_ylabel('$v \\ \\left( \\frac{ft^3}{h} \\right)$')
ax1.grid(which='BOTH', ls=':')
ax1.grid(True)

fig.tight_layout()
plt.show()
```
