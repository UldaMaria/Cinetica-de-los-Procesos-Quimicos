###  Producción de 300 millones de libras al año de etileno en un PFR: diseño de un PFR a escala industrial

Primero importaremos la librerías que nos ayudaran a resolver el problema (No olvides activar las librerías con **Shift+enter**):

```python
import numpy
import matplotlib.pyplot as plt
```

```python
CA0 = 1 * 6 / (0.73 * 1980)  # lbmol/ft3
FA0 = 0.34 / 0.8
eps = 1
k = 0.072 * numpy.exp(82000 * (1100 - 1000)/(1.987*1000*1100))
X = numpy.linspace(0, 0.8, 500)
V = (FA0 / (k*CA0)) * ((1+eps)*numpy.log(1/(1-X))-eps*X)  # ft3
z = V / (0.0205*100)  # ft
C_A = CA0 * (1-X) / (1 + eps*X)
C_C = CA0 * X / (1 + eps*X)
```

```python
# Graficando resultados
fig, ax = plt.subplots()
ax.minorticks_on()
ax.plot(z, C_A, 'k', linewidth=1.1, label='$C_A$')
ax.plot(z, C_C, 'k-', linewidth=0.6, label='$C_C, C_B$')
ax.set_xlabel('$Distancia\\ a \\ lo \\ largo \\ del \\ reactor \\ z(ft)$')
ax.set_ylabel('$Concentración \\ \\left( \\frac{lbmol}{ft^3} \\right)$')
ax.legend(loc='lower center', ncol=2)
ax.grid(which='BOTH', ls=':')
ax.grid(True)

ax1 = ax.twinx()
ax1.minorticks_on()
ax1.plot(z, X, 'k:', linewidth=1.5, label='$X$')
ax1.set_ylabel('$Conversión$')
ax1.legend(loc='upper center')

fig.tight_layout()

plt.show()
```
