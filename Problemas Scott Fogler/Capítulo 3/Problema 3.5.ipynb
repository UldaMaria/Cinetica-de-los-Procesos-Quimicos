Primero importaremos la librerías que nos ayudaran a resolver el problema (No olvides activar las librerías con **Shift+enter**):

```python
import matplotlib.pyplot as plt
import numpy
```

```python
# Datos
epsilon = 0.28 * (1 - 1 - 1/2)
C_SO2_0 = 0.28 * 1495 / (8.314 * 500)
thetaB = 0.72 * 0.21 / 0.28
thetaI = 0.72 * 0.70 / 0.28
X = numpy.arange(0.0, 1.0, 0.01)
C_SO2 = C_SO2_0 * (1 - X)/(1 + epsilon*X)
C_O2 = C_SO2_0 * (thetaB - 0.5*X) / (1 + epsilon*X)
C_SO3 = C_SO2_0 * X / (1 + epsilon*X)
C_N2 = C_SO2_0 * thetaI / (1 + epsilon*X)
```

```python
# Gráfica
fig = plt.figure()
ax = fig.add_subplot(111)
ax.minorticks_on()
ax.set_xlabel('$Conversión, \\ X$')
ax.set_ylabel('$Concentración \\ \\left( \\frac{gmol}{dm^3} \\right)$')
ax.set_xlim(left=0.0, right=1.0)
concentracionesLabel = ('$C_{SO_2}$', '$C_{O_2}$', '$C_{SO_3}$', '$C_{N_2}$')
concentracionesData = (C_SO2, C_O2, C_SO3, C_N2)
for i in range(len(concentracionesLabel)):
    ax.plot(X, concentracionesData[i], label=concentracionesLabel[i])
ax.grid(which='BOTH', ls=':')
ax.grid(True)
ax.legend()
plt.show()
```
