## Condicionales

Cuando existen determinadas condiciones y ella es verdadera entonces se denomina `if`. 

En el siguiente ejemplo, tenemos una secuencia de `if...elif...else`, que juntos proporcionan tres caminos diferentes para la ejecuci贸n. Si la primera condici贸n es falsa, se prueba la siguiente condici贸n elif; si este tambi茅n es falso, pasamos a else y lo ejecutamos. Intente cambiar el valor de `x` y vea qu茅 sucede cuando ejecuta la celda en cada caso.

```python
x = 7

if x > 7:
    print("x es mayor que 7")
elif x < 7: #note that Python contracts else if to elif
    print("x es menor que 7")
else:
    print("x es igual a 7")
```

### Colores

Para cambiar el color de los objetos se utiliza el par谩metro color = nombre-color, donde nombre-color es una cadena con el nombre del color de entre los colores disponibles.

```python
import matplotlib.pyplot as color
fig, ax = color.subplots()
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'], color = 'tab:purple')
ax.plot(dias, temperaturas['Mixco'], color = 'tab:green')
color.show()
```

### Marcadores 

Para cambiar la forma de los puntos marcadores se utiliza el par谩metro marker = nombre-marcador donde nombre-marcador es una cadena con el nombre del marcador de entre los marcadores disponibles

```python
import matplotlib.pyplot as marcador
fig, ax = marcador.subplots()
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'], marker = '^')
ax.plot(dias, temperaturas['Mixco'], marker = 'o')
marcador.show()
```

### Estilo de L铆neas

Para cambiar el estilo de las l铆neas se utiliza el par谩metro linestyle = nombre-estilo donde nombre-estilo es una cadena con el nombre del estilo de entre los estilos disponibles

```python
import matplotlib.pyplot as estilo_lineas
fig, ax = estilo_lineas.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'], linestyle = 'dashed')
ax.plot(dias, temperaturas['Mixco'], linestyle = 'dotted')
estilo_lineas.show()
```

### T铆tulos

ax.set_title(titulo, loc=alineacion, fontdict=fuente) : A帽ade un t铆tulo con el contenido de la cadena titulo a los ejes ax. El par谩metro loc indica la alineaci贸n del t铆tulo, que puede ser 'left' (izquierda), 'center' (centro) o 'right' (derecha), y el par谩metro fontdict indica mediante un diccionario las caracter铆sticas de la fuente (la el tama帽o fontisize, el grosor fontweight o el color color).

```python
import matplotlib.pyplot as ejemplo_T
fig, ax = ejemplo_T.subplots()
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'])
ax.plot(dias, temperaturas['Mixco'])
ax.set_title('Evoluci贸n de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ejemplo_T.show()
```

### Ejes

Para cambiar el aspecto de los ejes se suelen utilizar los siguientes m茅todos:

- **ax.set_xlabel(titulo) :** A帽ade un t铆tulo con el contenido de la cadena titulo al eje x de ax. Se puede personalizar la alineaci贸n y la fuente con los mismos par谩metros que para el t铆tulo principal.

- **ax.set_ylabel(titulo) :** A帽ade un t铆tulo con el contenido de la cadena titulo al eje y de ax. Se puede personalizar la alineaci贸n y la fuente con los mismos par谩metros que para el t铆tulo principal.

- **ax.set_xlim([limite-inferior, limite-superior]) :** Establece los l铆mites que se muestran en el eje x de ax.

- **ax.set_ylim([limite-inferior, limite-superior]) :** Establece los l铆mites que se muestran en el eje y de ax.

- **ax.set_xticks(marcas) :** Dibuja marcas en el eje x de ax en las posiciones indicadas en la lista marcas.

- **ax.set_yticks(marcas) :** Dibuja marcas en el eje y de ax en las posiciones indicadas en la lista marcas.

- **ax.set_xscale(escala) :** Establece la escala del eje x de ax, donde el par谩metro escala puede ser 'linear' (lineal) o 'log' (logar铆tmica).

- **ax.set_yscale(escala) :** Establece la escala del eje y de ax, donde el par谩metro escala puede ser 'linear' (lineal) o 'log' (logar铆tmica).

```python
import matplotlib.pyplot as ejes
fig, ax = ejes.subplots()
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'])
ax.plot(dias, temperaturas['Mixco'])
ax.set_xlabel("D铆as", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:red'})
ax.set_ylabel("Temperatura 潞C")
ax.set_ylim([20,35])
ax.set_yticks(range(20, 35))
ejes.show()
```

### Leyenda

Para a帽adir una leyenda a un gr谩fico se utiliza el siguiente m茅todo:

ax.legend(leyendas, loc = posici贸n) : Dibuja un leyenda en los ejes ax con los nombres indicados en la lista leyendas. El par谩metro loc indica la posici贸n en la que se dibuja la leyenda y puede ser 'upper left' (arriba izquierda), 'upper center' (arriba centro), 'upper right' (arriba derecha), 'center left' (centro izquierda), 'center' (centro), 'center right' (centro derecha), 'lower left' (abajo izquierda), 'lower center' (abajo centro), 'lower right' (abajo derecha). Se puede omitir la lista leyendas si se indica la leyenda de cada serie en la funci贸n que la dibuja mediante el par谩metro label.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Guatemala'], label = 'Guatemala')
ax.plot(dias, temperaturas['Mixco'], label = 'Mixco')
ax.legend(loc = 'upper right')
plt.show()
```

### Gr谩ficos m煤ltiples

Es posible dibujar varios gr谩ficos en distintos ejes en una misma figura organizados en forma de tabla. Para ello, cuando se inicializa la figura y los ejes, hay que pasarle a la funci贸n subplots el n煤mero de filas y columnas de la tabla que contendr谩 los gr谩ficos. Con esto los distintos ejes se organizan en un array y se puede acceder a cada uno de ellos a trav茅s de sus 铆ndices. Si se quiere que los distintos ejes compartan los mismos l铆mites para los ejes se pueden pasar los par谩metros sharex = True para el eje x o sharey = True para el eje y.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2, sharey = True)
dias = ['L', 'M', 'Mi', 'J', 'V', 'S', 'D']
temperaturas = {'Guatemala':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Mixco':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax[0, 0].plot(dias, temperaturas['Guatemala'])
ax[0, 1].plot(dias, temperaturas['Mixco'], color = 'tab:green')
ax[1, 0].bar(dias, temperaturas['Guatemala'])
ax[1, 1].bar(dias, temperaturas['Mixco'], color = 'tab:red')
plt.show()
```
